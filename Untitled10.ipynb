```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Sample frequency and time vector
fs = 1000  # 1kHz sampling
t = np.linspace(0, 1.0, fs)
# A signal with 5Hz, 50Hz, and 150Hz components
data = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*50*t) + np.sin(2*np.pi*150*t)

# --- 1. Low-Pass Filter (Removes high frequencies) ---
b_low, a_low = signal.butter(4, 20, btype='low', fs=fs)
low_passed = signal.filtfilt(b_low, a_low, data)

# --- 2. High-Pass Filter (Removes low frequencies) ---
b_high, a_high = signal.butter(4, 100, btype='high', fs=fs)
high_passed = signal.filtfilt(b_high, a_high, data)

# --- 3. Band-Pass Filter (Allows a specific range) ---
b_band, a_band = signal.butter(4, [40, 60], btype='band', fs=fs)
band_passed = signal.filtfilt(b_band, a_band, data)

# Plotting Results
plt.figure(figsize=(12, 8))
plt.subplot(4,1,1); plt.plot(t, data); plt.title('Original Signal')
plt.subplot(4,1,2); plt.plot(t, low_passed); plt.title('Low-Pass Filter (Cutoff 20Hz)')
plt.subplot(4,1,3); plt.plot(t, high_passed); plt.title('High-Pass Filter (Cutoff 100Hz)')
plt.subplot(4,1,4); plt.plot(t, band_passed); plt.title('Band-Pass Filter (40Hz-60Hz)')
plt.tight_layout()
plt.show()
```
